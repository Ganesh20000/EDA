```python
df= bios.copy()
```

```python

#^ getting rid of .
df['Name']=df['Used name'].str.replace('•',' ')
```

```python
df.tail()
```

```python

#! split height and weight on /
measure = df['Measurements'].str.split('/',)

df['height']= measure.str.get(0).str.replace(r'\D',' ',regex=True)

df['weight']=measure.str.get(1).str.replace(r'\D', ' ',regex=True)
```

```python
df[['weight','height']].sample(54)
```

```python

#* dropping the measurement column
df.drop(columns={'Measurements'},inplace=True)
```

```python
df.columns
```

```python

#! parse out date form born and dead column
date_pattern = r'(\d+ \w+ \d{4})'
df['Born_date'] =df['Born'].str.extract(r'(\d+ \w+ \d{4})')
```

```python
df['Died_date']=df['Died'].str.extract(date_pattern)
```

```python
df[['Born_date','Died_date']].sample(6)
```

```python


#! now finding the reverse thing means checking if date_format 
#^ is not follwoing or notna value

df[~df['Born'].str.match(date_pattern,na=False) & df['Born'].notna()]
```
